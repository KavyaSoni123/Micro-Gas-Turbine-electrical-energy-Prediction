```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../Micro-Gas-Turbine-electrical-energy-Prediction/micro+gas+turbine+electrical+energy+prediction/train/ex_1.csv')
```

```python
df
```

```python
folder_path = '../Micro-Gas-Turbine-electrical-energy-Prediction/micro+gas+turbine+electrical+energy+prediction/train/'
csv_files = []
for file_name in os.listdir(folder_path):
    if file_name.endswith('.csv'):
        file_path = os.path.join(folder_path, file_name)
        df = pd.read_csv(file_path)
        csv_files.append(df)

merged_df = pd.concat(csv_files, ignore_index=True)
```

```python
merged_df
```

```python
merged_df['time'] = merged_df['time'].round(2)
```

```python
merged_df
```

```python
merged_df.to_csv('merged_output_train.csv', index=False)
```
